```python
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
def make_graph(stock_data, revenue_data, stock):
    fig = make_subplots(rows=2, cols=1, shared_xaxes=True, subplot_titles=("Historical Share Price", "Historical Revenue"), vertical_spacing = .3)
    fig.add_trace(go.Scatter(x=pd.to_datetime(stock_data.Date, infer_datetime_format=True), y=stock_data.Close.astype("float"), name="Share Price"), row=1, col=1)
    fig.add_trace(go.Scatter(x=pd.to_datetime(revenue_data.Date, infer_datetime_format=True), y=revenue_data.Revenue.astype("float"), name="Revenue"), row=2, col=1)
    fig.update_xaxes(title_text="Date", row=1, col=1)
    fig.update_xaxes(title_text="Date", row=2, col=1)
    fig.update_yaxes(title_text="Price ($US)", row=1, col=1)
    fig.update_yaxes(title_text="Revenue ($US Millions)", row=2, col=1)
    fig.update_layout(showlegend=False,
    height=900,
    title=stock,
    xaxis_rangeslider_visible=True)
    fig.show()
```

```python
import pandas as pd
import requests
from io import StringIO

url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"
headers = {"User-Agent": "Mozilla/5.0"}

# Get HTML and parse with StringIO to avoid warning
response = requests.get(url, headers=headers)
html_content = StringIO(response.text)
tables = pd.read_html(html_content)

# Find quarterly revenue table (typically the 2nd table)
print(f"Found {len(tables)} tables")  # Debug table count
gme_revenue = tables[1]  # Try different indices if needed

# Clean and format data
gme_revenue.columns = ['Date', 'Revenue']
gme_revenue['Revenue'] = gme_revenue['Revenue'].str.replace(r'[$,]', '', regex=True)
gme_revenue['Revenue'] = pd.to_numeric(gme_revenue['Revenue'], errors='coerce')
gme_revenue = gme_revenue.dropna().reset_index(drop=True)

print(gme_revenue.tail())
```

```python
import pandas as pd
import yfinance as yf
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Fetch historical data
tesla = yf.Ticker("TSLA")
tsla_df = tesla.history(start="2012-01-01").reset_index()

def make_graph(stock_data):
    fig = make_subplots(rows=2, cols=1, shared_xaxes=True, vertical_spacing=0.1)
    
    # Price Chart
    fig.add_trace(go.Candlestick(
        x=stock_data['Date'],
        open=stock_data['Open'],
        high=stock_data['High'],
        low=stock_data['Low'],
        close=stock_data['Close'],
        name='Price'
    ), row=1, col=1)
    
    # Volume Chart
    fig.add_trace(go.Bar(
        x=stock_data['Date'],
        y=stock_data['Volume'],
        name='Volume',
        marker_color='blue'
    ), row=2, col=1)

    # Layout Configuration
    fig.update_layout(
        title='Tesla Stock Price and Volume History (2012-Present)',
        xaxis=dict(
            rangeselector=dict(
                buttons=list([
                    dict(count=1, label="1m", step="month"),
                    dict(count=6, label="6m", step="month"),
                    dict(count=1, label="YTD", step="year"),
                    dict(count=1, label="1y", step="year"),
                    dict(step="all")
                ])
            ),
            rangeslider=dict(visible=True),
            type="date"
        ),
        yaxis_title='Price (USD)',
        yaxis2_title='Volume',
        height=600
    )
    
    return fig

# Generate and display
fig = make_graph(tsla_df)
fig.show()
```

```python
import pandas as pd
import yfinance as yf
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.io as pio

# Set renderer to browser if needed (uncomment below line)
# pio.renderers.default = "browser"

# Fetch GameStop historical data
gme = yf.Ticker("GME")
gme_df = gme.history(start="2000-01-01").reset_index()  # Extra early start date

def make_graph(stock_data):
    # Create figure with volume subplot
    fig = make_subplots(rows=2, cols=1, 
                       shared_xaxes=True,
                       vertical_spacing=0.1,
                       row_heights=[0.7, 0.3])
    
    # Candlestick Chart
    fig.add_trace(go.Candlestick(
        x=stock_data['Date'],
        open=stock_data['Open'],
        high=stock_data['High'],
        low=stock_data['Low'],
        close=stock_data['Close'],
        name='Price'
    ), row=1, col=1)
    
    # Volume Bar Chart
    fig.add_trace(go.Bar(
        x=stock_data['Date'],
        y=stock_data['Volume'],
        name='Volume',
        marker_color='purple'  # Different color for GME
    ), row=2, col=1)

    # Configure layout
    fig.update_layout(
        title='GameStop Stock Price and Volume History (2006-Present)',
        xaxis=dict(
            rangeselector=dict(
                buttons=list([
                    dict(count=1, label="1m", step="month"),
                    dict(count=6, label="6m", step="month"),
                    dict(count=1, label="YTD", step="year"),
                    dict(count=1, label="1y", step="year"),
                    dict(step="all")
                ])
            ),
            rangeslider=dict(visible=True),
            type="date"
        ),
        yaxis_title='Price (USD)',
        yaxis2_title='Volume',
        height=600,
        template='plotly_white'
    )
    
    # Set x-axis range to start from 2006
    fig.update_xaxes(range=['2006-01-01', stock_data['Date'].max()])
    
    return fig

# Generate and display graph
fig = make_graph(gme_df)
fig.show()
```
